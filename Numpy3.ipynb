# Maths Operations 

```python
x= [21,31,61,91] 
y=[101,95,17,13] 
a=[11,13,17,19]
b= [21,13,12,17]
```

```python
check = np.array([x,y,a,b])
check
```

```python
# sum() 
np.sum(check) # returns the sum of all item in the array 
```

```python
np.sum(check, axis=0) # find the sum column wise 
```

```python
np.sum(check, axis=1) # find sum row wise 
```

```python
# mean() 
np.mean(check) # finds mean of the items in array 
```

```python
np.mean(check , axis=0) # find mean column wise
```

```python
np.mean(check, axis=1) # finds mean row wise 
```
